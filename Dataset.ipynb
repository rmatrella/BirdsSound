```python
import os
import pyogg
import soundfile as sf
import librosa
from pydub import AudioSegment
from pathlib import Path

entries = os.listdir('drive/MyDrive/CIDL project/birds/train_audio')
# AudioSegment.ffmpeg = 'C:/ffmpeg/bin/ffmpeg.exe'
# AudioSegment.ffprobe = 'C:/ffmpeg/bin/ffprobe.exe'
# AudioSegment.ffplay = 'C:/ffmpeg/bin/ffplay.exe'

for entry in entries: #for each specie (one folder for each one)
    print(entry)
    files = os.listdir('drive/MyDrive/CIDL project/birds/train_audio/' + entry)
    #create directory
    try:
        os.makedirs("drive/MyDrive/CIDL project/birds/converted/" + entry)
        print("Folder %s created!" % entry)
    except FileExistsError:
        print("Folder %s already exists" % entry)
    
    #get all the audio for each specie   
    for file in files: 
        
        name = file.split('.')[0]
        myfile = Path('drive', 'MyDrive', 'CIDL project', 'birds', 'train_audio', entry, file)
        
        print(myfile)
        #convert audio 
        audio = AudioSegment.from_file(myfile)
        audio.export("drive/MyDrive/CIDL project/birds/converted/"+entry+"/"+name+".wav", format="wav")
```

```python
# get minutes of audio for each specie 
```

```python
import os
import pyogg
import soundfile as sf
import librosa
from pydub import AudioSegment
from pathlib import Path

dictionary = {}
entries = os.listdir('drive/MyDrive/CIDL project/birds/train_audio/')


for entry in entries: #for each specie (one folder for each one)
    files = os.listdir('drive/MyDrive/CIDL project/birds/train_audio/' + entry)
    time = 0
    
    #get all the audio for each specie   
    for file in files: 
        
        path = 'drive/MyDrive/CIDL project/birds/train_audio/' + entry + '/' + file
        sound, sr = librosa.load(path)
        time += librosa.get_duration(y=sound, sr=sr)
        
    print(entry, time/60)   
    dictionary[entry] = time/60
    
```

```python
# Order items inside dictionary 
```

```python
d = dict(sorted(dictionary.items(), key=lambda item: item[1], reverse=True)[:150])
```

# Get common names of directory label 

```python
import csv
dictionary = {}

with open('drive/MyDrive/CIDL project/birds/train_metadata.csv', mode='r', encoding="utf8") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            print(f'\t{row[0]} to change into {row[6]}.')
            if row[0] not in dictionary:
                dictionary[row[0]] = row[6]
            line_count += 1
    print(f'Processed {line_count} lines.')
print(dictionary)
```

# Rename directories with common names

```python
import os 
os.chdir(r"drive/MyDrive/CIDL project/birds/converted")

for key, value in dictionary.items():
    os.rename(key,dictionary[key])
```
